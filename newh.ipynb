# Task 1: Compare Server Headers

```python
import requests

naver_response = requests.get("https://www.naver.com")
print("Naver")
for key, value in naver_response.headers.items():
    print(f"{key}: {value}")

print("\n")

google_response = requests.get("https://www.google.com")
print("Google")
for key, value in google_response.headers.items():
    print(f"{key}: {value}")

print("\n")
compare_keys = ['Server', 'Content-Type', 'Content-Encoding']
for key in compare_keys:
    naver_val = naver_response.headers.get(key, " ")
    google_val = google_response.headers.get(key, " ")
    print(f"\n[{key}]")
    print(f"  네이버: {naver_val}")
    print(f"  구글:   {google_val}")
```

# Task 4: API + Gradio UI

```python
import requests
import pandas as pd
import gradio as gr

def filter_by_userid(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {"userId": int(user_id)}
    response = requests.get(url, params=params)
    data = response.json()

    if not data:
        return pd.DataFrame({"Result": ["No data for this userId."]})

    df = pd.DataFrame(data)
    return df

demo = gr.Interface(
    fn=filter_by_userid,
    inputs=gr.Number(label="Enter userId (1-10)", precision=0),
    outputs="dataframe",
    title="Network & Web Data Collection Practice",
    description="Enter a userId to retrieve data for that user and display it as a Pandas DataFrame."
)

print("=" * 60)
print("Gradio UI is running!")
print("Click the link below to view the web data collection result.")
print("=" * 60)

demo.launch()
```

# Conclusion

The application successfully retrieves data from the API and dynamically filters it based on the user-provided userId.
By integrating Gradio, the system allows real-time interaction, and the filtered results are displayed in a Pandas DataFrame.
This demonstrates how to combine APIs, data processing, and user interfaces for interactive data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gradio as gr
import os


def read_csv_flexible(path):
    encodings = ["utf-8", "cp949", "euc-kr"]
    for enc in encodings:
        try:
            return pd.read_csv(path, encoding=enc)
        except:
            continue
    raise ValueError(f"Failed: {path}")
```

```python
csv_files = [f for f in os.listdir() if f.endswith(".csv")]
csv_files
```

```python
dfs = []
for f in csv_files:
    df = read_csv_flexible(f)
    df["source_file"] = f
    dfs.append(df)

len(dfs)
```

# Task 6: Gemini API (AI Integration)

```python
import google.generativeai as genai
import gradio as gr
```

```python
genai.configure(api_key="YOUR_API_KEY")
```

```python
model = genai.GenerativeModel("gemini-1.5-flash-latest")
```

```python
def translate_first_line(text):
    try:
        if not text or not text.strip():
            return "Please enter text."

        first_line = text.strip().split("\n")[0]

        prompt = f"Translate this Korean text into English:\n{first_line}"

        response = model.generate_content(prompt)

        return response.text

    except Exception as e:
        return f"Error: {e}"
```

```python
demo = gr.Interface(
    fn=translate_first_line,
    inputs=gr.Textbox(lines=4, label="Enter Task 7 Text"),
    outputs=gr.Textbox(label="Translation"),
    title="Task 9 - Gemini Translator",
    description="Translate only the first line of Task 7."
)
```

```python
demo.launch()
```
